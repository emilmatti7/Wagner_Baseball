# Wagner Baseball - Analytics Dashboard
This python notebook is to scrape the HTML table elements off of Wagner Baseball's website and turn it into a dataframe/csv for further analysis in SQL. After manipulation, a dashboard will be made in Tableau.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# find html element <table>
df21 = pd.read_html('https://wagnerathletics.com/sports/baseball/stats/2021')
df20 = pd.read_html('https://wagnerathletics.com/sports/baseball/stats/2020')
df19 = pd.read_html('https://wagnerathletics.com/sports/baseball/stats/2019')
df18 = pd.read_html('https://wagnerathletics.com/sports/baseball/stats/2018')
df17 = pd.read_html('https://wagnerathletics.com/sports/baseball/stats/2017')
df16 = pd.read_html('https://wagnerathletics.com/sports/baseball/stats/2016')
df15 = pd.read_html('https://wagnerathletics.com/sports/baseball/stats/2015')
df14 = pd.read_html('https://wagnerathletics.com/sports/baseball/stats/2014')
```

```python
# team pitching stats
pitching_stats2021 = df21[7]
pitching_stats2020 = df20[7]
pitching_stats2019 = df19[7]
pitching_stats2018 = df18[7]
pitching_stats2017 = df17[7]
pitching_stats2016 = df16[7]
pitching_stats2015 = df15[7]
pitching_stats2014 = df14[7]
```

```python
# individual pitching stats
indi_pitching_stats2021 = df21[1]
indi_pitching_stats2020 = df20[1]
indi_pitching_stats2019 = df19[1]
indi_pitching_stats2018 = df18[1]
indi_pitching_stats2017 = df17[1]
indi_pitching_stats2016 = df16[1]
indi_pitching_stats2015 = df15[1]
indi_pitching_stats2014 = df14[1]
```

```python
# append all pitching stats
pitch_stats = pd.concat([pitching_stats2021,pitching_stats2020,pitching_stats2019, pitching_stats2018, pitching_stats2017,pitching_stats2016,pitching_stats2015, pitching_stats2014], ignore_index = True)
indi_pitch_stats = pd.concat([indi_pitching_stats2021, indi_pitching_stats2020, indi_pitching_stats2019, indi_pitching_stats2018, indi_pitching_stats2017, indi_pitching_stats2016, indi_pitching_stats2015, indi_pitching_stats2014], ignore_index = True)
```

```python
# individual hitting stats
hitting_stats2021 = df21[0]
hitting_stats2020 = df20[0]
hitting_stats2019 = df19[0]
hitting_stats2018 = df18[0]
hitting_stats2017 = df17[0]
hitting_stats2016 = df16[0]
hitting_stats2015 = df15[0]
hitting_stats2014 = df14[0]
```

```python
# append all hitting stats
hitting_stats = pd.concat([hitting_stats2021,hitting_stats2020,hitting_stats2019, hitting_stats2018, hitting_stats2017,hitting_stats2016,hitting_stats2015, hitting_stats2014], ignore_index = True)
```

```python
# hitting stats
game_stats2021 = df21[6]
game_stats2020 = df20[6]
game_stats2019 = df19[6]
game_stats2018 = df18[6]
game_stats2017 = df17[6]
game_stats2016 = df16[6]
game_stats2015 = df15[6]
game_stats2014 = df14[6]
```

```python
# append all pitching game
game_stats = pd.concat([game_stats2021,game_stats2020,game_stats2019, game_stats2018, game_stats2017,game_stats2016,game_stats2015, game_stats2014], ignore_index = True)
```

```python
# convert these dataframes to a csv
game_stats.to_csv('game_stats.csv')
hitting_stats.to_csv('hitting_stats.csv')
pitch_stats.to_csv('indi_pitch_stats.csv')
```
